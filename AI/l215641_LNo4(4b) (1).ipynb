```python
# Function to read the cube from a text file
def read_cube(filename):
    cube = []
    with open(filename, 'r') as file:
        for line in file:
            row = line.split()
            cube.append(row)
    return cube
def checkcount(x,y,cube):
     count=0
     if 0<=y+1< len(cube[0]) and cube[x][y+1]=='0':
            count+=1
     if 0<=x+1< len(cube) and cube[x+1][y]=='0':
            count+=1
     if 0<=x-1< len(cube) and cube[x-1][y]=='0':
            count +=1
     if 0<=y-1< len(cube[0]) and cube[x][y-1]=='0':
            count +=1
     return count
# Function to implement Depth-First Search (DFS)
def dfs(cube):
    moves=0
    stack = [(0, 0)]  # Starting point at (0, 0)
    visited = set()
    path = []
    cflag=0
    while stack:
        x, y = stack.pop()
        if cube[x][y] == 'G':  # Goal state found
                moves +=1
                print("Total number of moves use in dfs to reach goal : ",moves)
                path.append((x, y))
                return path
        visited.add((x, y))
        path.append((x, y))
        if 0 <= x+1 < len(cube) and 0 <= y < len(cube[0])  and (x+1, y) not in visited :
              if cube[x+1][y]=='0' or cube[x+1][y]=='G':
                  moves +=1
                  stack.append((x+1, y))  # Move down
                  cflag=1

        if 0 <= x < len(cube) and 0 <= y+1 < len(cube[0]) and  (x, y+1) not in visited :
              if cube[x][y+1]=='0' or cube[x][y+1]=='G':
                  moves +=1
                  stack.append((x, y+1))  # Move right
                  cflag=1

        if 0 <= x-1 < len(cube) and 0 <= y < len(cube[0]) and  (x-1, y) not in visited :
              if cube[x-1][y]=='0' or cube[x-1][y]=='G':
                  moves +=1
                  stack.append((x-1, y))  # Move up
                  cflag=1

        if 0 <= x < len(cube) and 0 <= y-1 < len(cube[0]) and (x, y-1) not in visited :
              if cube[x][y-1]=='0' or cube[x][y-1]=='G':
                  moves +=1
                  stack.append((x, y-1))  # Move left
                  cflag=1
        if cflag==0:
              nx,ny=stack.pop()
              x,y=path.pop()
              while (nx,ny)!=(x+1,y) and (nx,ny)!=(x,y-1) and (nx,ny)!=(x-1,y) and (nx,ny)!=(x,y+1):
                     x,y=path.pop()
              stack.append((nx,ny))
              path.append((x,y))
        cflag=0
    print("Total number of wasted moves use in dfs : ",moves)
    return -1





# Function to implement Breadth-First Search (BFS)
def bfs(cube):
    moves=0
    queue = [(0, 0)]  # Starting point at (0, 0)
    visited = set()
    path = []

    while queue:
        x, y = queue.pop(0)
        if cube[x][y] == 'G':  # Goal state found
                moves +=1
                print("Total number of moves use in bfs to reach goal : ",moves)
                path.append((x, y))
                return path
        visited.add((x, y))
        path.append((x, y))
        if 0 <= x+1 < len(cube) and 0 <= y < len(cube[0])  and (x+1, y) not in visited :
              if cube[x+1][y]=='0' or cube[x+1][y]=='G':
                  moves +=1
                  queue.append((x+1, y))  # Move down
                  cflag=1

        if 0 <= x < len(cube) and 0 <= y+1 < len(cube[0]) and  (x, y+1) not in visited :
              if cube[x][y+1]=='0' or cube[x][y+1]=='G':
                  moves +=1
                  queue.append((x, y+1))  # Move right
                  cflag=1

        if 0 <= x-1 < len(cube) and 0 <= y < len(cube[0]) and  (x-1, y) not in visited :
              if cube[x-1][y]=='0' or cube[x-1][y]=='G':
                  moves +=1
                  queue.append((x-1, y))  # Move up
                  cflag=1

        if 0 <= x < len(cube) and 0 <= y-1 < len(cube[0]) and (x, y-1) not in visited :
              if cube[x][y-1]=='0' or cube[x][y-1]=='G':
                  moves +=1
                  queue.append((x, y-1))  # Move left
                  cflag=1
        if cflag==0:
              (x,y)=path.pop()
              removelist=[(x,y)]
              while 1:
                  if 0<=y+1< len(cube[0]) and cube[x][y+1]=='0'and (x,y+1) not in removelist:
                      y=y+1
                      removelist.append((x,y))
                  elif 0<=x+1< len(cube) and cube[x+1][y]=='0'and (x+1,y) not in removelist:
                      x=x+1
                      removelist.append((x,y))
                  elif 0<=x-1< len(cube) and cube[x-1][y]=='0'and (x-1,y) not in removelist:
                      x=x-1
                      removelist.append((x,y))
                  elif 0<=y-1< len(cube[0]) and cube[x][y-1]=='0'and (x,y-1) not in removelist:
                       y=y-1
                       removelist.append((x,y))
                  count=checkcount(x,y,cube)
                  if count < 3:
                     path.remove((x,y))
                  else:
                    break
        cflag=0
    print("Total number of wasted moves use in bfs : ",moves)
    return -1

# Reading the cube from a text file
cube = read_cube('/cube.txt')

# Implementing DFS and BFS on the cube
dfs_path = dfs(cube)
print("DFS Path:", dfs_path)

print("\n")

bfs_path = bfs(cube)
print("BFS Path:", bfs_path)
```

```python
# Function to read the cube from a text file
def read_cube_from_file(filename):
    cube = []
    with open(filename, 'r') as file:
        for line in file:
            row = line.split()
            cube.append(row)
    return cube

# Function to implement A* Search
def astar(cube):
    open_list = []
    closed_list = set()
    start_state = (0, 0)
    goal_state = (len(cube)-1, len(cube[0])-1)
    heuristic = lambda state: abs(state[0] - goal_state[0]) + abs(state[1] - goal_state[1])

    open_list.append((heuristic(start_state), start_state))

    while open_list:
        current_state = open_list.pop(0)
        if current_state[1] == goal_state:
            return current_state[1]
        closed_list.add(current_state[1])

        neighbors = [(current_state[1][0]+1, current_state[1][1]), (current_state[1][0], current_state[1][1]+1),
                     (current_state[1][0]-1, current_state[1][1]), (current_state[1][0], current_state[1][1]-1)]

        for neighbor in neighbors:
            if neighbor not in closed_list:
                if 0 <= neighbor[0] < len(cube) and 0 <= neighbor[1] < len(cube[0]):
                    if cube[neighbor[0]][neighbor[1]] == 0 or cube[neighbor[0]][neighbor[1]] == 2:
                        new_state = (neighbor[0], neighbor[1])
                        new_heuristic = heuristic(new_state)
                        new_cost = current_state[0] + 1
                        if new_state not in closed_list or new_cost < open_list[-1][0]:
                            open_list.append((new_heuristic + new_cost, new_state))
                            closed_list.add(new_state)

    return -1

# Reading the cube from a text file
cube = read_cube('cube.txt')

# Implementing A* Search on the cube
path = astar(cube)

print("A* Path:", path)
```
