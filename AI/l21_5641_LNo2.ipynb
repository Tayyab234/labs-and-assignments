```python
#task12
def binary_search(nums, target):
    low = 0
    high = len(nums) - 1

    while low <= high:
        mid = low + (high - low) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
nums = [1, 3, 5, 7, 9]
binary_search(nums, 3)
```

```python
#task13
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result
arr = [70, 50, 30, 10, 20, 40]
sorted_arr = merge_sort(arr)
print(sorted_arr)
```

```python
#task14
def quick_sort(arr):

  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]
  return quick_sort(left) + middle + quick_sort(right)

arr = [1, 7, 4, 1, 10, 9, -2]
sorted_arr = quick_sort(arr)
print(sorted_arr)
```

```python
#task15
def dictcount(l1):
  Dt=dict()
  for i in range(0,len(l1)):
    Dt[l1[i]]=0
  for i in range(0,len(l1)):
    Dt[l1[i]]+=1
  return Dt
l1=[1, 2, 2, 3, 4, 2]
print(dictcount(l1))
```

```python
#task21
def filterDict(dictionary,Ltkey):
  newdict=dict()
  for i in range(0,len(Ltkey)):
    newdict[Ltkey[i]]=dictionary[Ltkey[i]]
  return newdict
keys=["blue","red","green"]
dictionary={"blue":10,"black":12,"red":9,"green":7,"yellow" : 5}
print(filterDict(dictionary,keys))
```
