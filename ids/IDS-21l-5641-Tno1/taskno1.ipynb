```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('data.csv')
df.head()
```

```python
k=df.Yearexp.sum()#x
h=df.Yearexp
v=df.Salary
c=h*h
u=c.sum()  #x^2
c=df.Salary.sum()#y
n=5
d=k*k #(x)^2
q=(h*v).sum()#xy

slope=(n*q-k*c)/(n*u-d)
intercept=(u*c-k*q)/(n*u-d)

#y-predicted
print("Linear regression y-predicted values are : ")
y=slope*df.Yearexp+intercept
print(y)

```

```python
#Batch Gradient decent

def costfind(y1):       #return mse
    h=((df.Salary-y1)*(df.Salary-y1)).sum()
    h=h/5
    return h

def predict(m,b):        #return y peridicted value
    y1=m*df.Yearexp+b
    return y1

def direction1(y1):             #return derivative of slope
    c=(-2*df.Yearexp*(df.Salary-y1)).sum()
    c=c/5
    return c

def direction2(y1):              #return derivative of intercept
    c=(-2*(df.Salary-y1)).sum()
    c=c/5
    return c

def stepsize(h,lr,c):    #return stepsize new value of slope and intercept
    k=h-lr*c
    return k

m=0
b=0
lr=0.01
i=1
for i in range(3):
  y1=predict(m,b)
  mse=costfind(y1)
  m1=direction1(y1)
  b1=direction2(y1)
  m=stepsize(m,lr,m1)
  b=stepsize(b,lr,b1)
  print("After  ",i," iteration :  ","\nValue of slope is : ",m,"\nValue of intercept is : ",b)
```

```python
#Stocastic Gradient decent

def costfind(y1,i):       #return mse
    h=((df.Salary[i]-y1)*(df.Salary[i]-y1))
    return h

def predict(m,b,i):        #return y peridicted value
    y1=m*df.Yearexp[i]+b
    return y1

def direction1(y1,i):             #return derivative of slope
    c=(-2*df.Yearexp[i]*(df.Salary[i]-y1))
    return c

def direction2(y1,i):              #return derivative of intercept
    c=(-2*(df.Salary[i]-y1))
    return c

def stepsize(h,lr,c):    #return stepsize new value of slope and intercept
    k=h-lr*c
    return k

m=0
b=0
lr=0.01
i=1
for i in range(3):
    for j in  range(5):
      y1=predict(m,b,j)
      mse=costfind(y1,j)
      m1=direction1(y1,j)
      b1=direction2(y1,j)
      m=stepsize(m,lr,m1)
      b=stepsize(b,lr,b1)
      print(i,"  iteration After  ",j," pass :  ","\nValue of slope is : ",m,"\nValue of intercept is : ",b)

    print("_________________________________________________________________")
```

```python
#Mini-Batch Gradient decent
#batch size=2 learning rate =0.001

def costfind(y1,i):       #return mse
    k=0
    h=0
    while k<2:
      if i<5:  
        h=h+((df.Salary[i]-y1)*(df.Salary[i]-y1))
        i=i+1
      k=k+1
    
    h=h/i
    return h

def predict(m,b,i):        #return y peridicted value
    k=0
    y1=0
    while k<2:
      if i<5:  
         y1=y1+(m*df.Yearexp[i]+b)
      i=i+1
      k=k+1
    return y1

def direction1(y1,i):             #return derivative of slope
    k=0
    c=0
    while k<2:
      if i<5:  
        c=c+(-2*df.Yearexp[i]*(df.Salary[i]-y1))
        i=i+1
      k=k+1
    
    c=c/i
    return c

def direction2(y1,i):             #return derivative of intercept
    k=0
    c=0
    while k<2:
      if i<5:  
        c=c+(-2*(df.Salary[i]-y1))
        i=i+1
      k=k+1
    
    c=c/i
    return c

def stepsize(h,lr,c):    #return stepsize new value of slope and intercept
    k=h-lr*c
    return k

m=0
b=0
lr=0.001
i=1
for i in range(3):
    for j in  range(0,5,2):
      y1=predict(m,b,j)
      mse=costfind(y1,j)
      m1=direction1(y1,j)
      b1=direction2(y1,j)
      m=stepsize(m,lr,m1)
      b=stepsize(b,lr,b1)
      print(i,"  iteration After  ",j," pass :  ","\nValue of slope is : ",m,"\nValue of intercept is : ",b)

    print("_________________________________________________________________")
```
