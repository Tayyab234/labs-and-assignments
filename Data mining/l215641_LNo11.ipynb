```python
#Qusetion 1
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/OnlineRetail.csv')

# Preprocess the dataset by removing missing values and outliers
df = df.dropna()
df = df[df['Quantity'] > 0]

# Scale the data for better clustering
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Quantity', 'UnitPrice']])
# Try different number of clusters
for n_clusters in [2, 3, 4]:
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, max_iter=50)
    kmeans.fit(df_scaled)

    # Visualize the results
    plt.figure(figsize=(10, 7))
    plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=kmeans.labels_)
    plt.xlabel('Quantity')
    plt.ylabel('UnitPrice')
    plt.title('K-means clustering with {} clusters'.format(n_clusters))
    plt.show()
```

```python
#Question no 2
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/OnlineRetail.csv')
df=df.head(13000)
# Preprocess the dataset by removing missing values and outliers
df = df.dropna()
df = df[df['Quantity'] > 0]

# Scale the data for better clustering
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Quantity', 'UnitPrice']])

# Try different number of clusters
for n_clusters in [2, 3, 4]:
    # Perform Hierarchical clustering
    cluster = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
    cluster.fit_predict(df_scaled)

    # Visualize the results
    plt.figure(figsize=(10, 7))
    plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=cluster.labels_)
    plt.xlabel('Quantity')
    plt.ylabel('UnitPrice')
    plt.title('Hierarchical clustering with {} clusters'.format(n_clusters))
    plt.show()
```
