```python
import matplotlib.pyplot as plt

# Sample data for multiple students
students = ['Student 1', 'Student 2', 'Student 3']
math_scores = [90, 85, 78]
english_scores = [80, 92, 87]
science_scores = [75, 68, 90]

# Calculate the average scores for each subject
average_math_score = sum(math_scores) / len(math_scores)
average_english_score = sum(english_scores) / len(english_scores)
average_science_score = sum(science_scores) / len(science_scores)

# Subjects and corresponding average scores
subjects = ['Math', 'English', 'Science']
average_scores = [average_math_score, average_english_score, average_science_score]

# Create a pie chart to represent the average scores
plt.pie(average_scores, labels=subjects, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Average Student Performance by Subject')

# Display the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for multiple students
students = ['Student 1', 'Student 2', 'Student 3']
math_scores = [90, 85, 78]
english_scores = [80, 92, 87]
science_scores = [75, 68, 90]

# Create a 1x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
# Plot the pie charts for Math, English, and Science scores
axes[0][0].pie(math_scores, labels=students, autopct='%1.2f%%')
axes[0][0].set_title('Math score of Students')
axes[0][1].pie(science_scores, labels=students, autopct='%1.2f%%')
axes[0][1].set_title('Science score of Students')

axes[0][2].pie(english_scores, labels=students, autopct='%1.2f%%')
axes[0][2].set_title('English score of Students')
plt.show()


#fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharex=True, sharey=True)

# Remove the unwanted subplot
#fig.delaxes(axes[0])




```
