```python
import csv
import matplotlib.pyplot as plt
from collections import defaultdict
import pandas as pd
inp=int(input('1. sepal.length   2. sepal.width\n3. petal.length   4. petal.width\nPlease choose one option to draw bar graph with respect to variety : '))
f=''
if inp==1:
    f='sepal.length'
elif inp==2:
    f='sepal.width'
elif inp==3:
    f='petal.length'
else:
    f='petal.width'
data=pd.read_csv('iris.csv')
# Initialize a dictionary to store the average sepal length for each variety
sepal_length_by_variety = defaultdict(list)
for i in range(0,len(data)):
   variety = data['variety'][i]
   sepal_length = float(data[f][i])
   sepal_length_by_variety[variety].append(sepal_length)
name=[]
avg=[]
for variety in data['variety']:
   if variety not in name:
      k=len(sepal_length_by_variety[variety])
      h=sum(sepal_length_by_variety[variety])
      c=h/k
      avg.append(c)
      name.append(variety)
print(avg)
print(name)
fig,axes=plt.subplots(2,2,figsize=(25,25))
axes[0][0].bar(name,avg,width=0.72,color='c')
axes[0][0].set_xticklabels(name,rotation=30,fontsize=15)
axes[0][0].set_xlabel('varieties',fontsize=18)
axes[0][0].set_ylabel('average {0}'.format(f),fontsize=18)
axes[0][0].set_title('bar graph with respect to average {0}'.format(f),fontsize=22)

axes[0][1].pie(avg,labels=name,autopct='%1.2f%%')
axes[0][1].set_title('Pie graph with respect to average {0}'.format(f),fontsize=22)


axes[1][0].scatter(name,avg , color='m', s=200, marker='o')
axes[1][0].set_ylabel('Average {0}'.format(f), fontsize=18)
axes[1][0].set_xlabel('Varieties', fontsize=18)
axes[1][0].set_title('Scatter plot of {0} by Variety'.format(f), fontsize=22)

axes[1][1].set_title('Line graph of {0} by Variety'.format(f), fontsize=22)
axes[1][1].plot(name,avg,label=f)
axes[1][1].legend()
plt.show()
```
