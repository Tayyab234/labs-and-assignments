**Part I: Find the matching substrings (Marks=2)** <br>



Write a function get_substring_positions() that accepts two strings as arguments. Return the **number of the positions** where they contain the same substring of length 2. For example, **"welldone"** and **"wearedone"** should yield 1.

````python
#### Write your code here ####
#return index position of s1
def  get_substring_positions(s1,s2):
     k=0
     for i in range(0,len(s1)-2):
            for j in range(0,len(s2)-2):
                 if s1[i:i+2]==s2[j:j+2]:
                    k=i
     return k
````

**Note:** You are not allowed to use Loop (of any type)
